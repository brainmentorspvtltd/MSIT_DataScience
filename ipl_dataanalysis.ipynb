# Data Visualization

```python
plt.figure(figsize=(10,7))
sns.countplot(x=matches['season'], data=matches)
plt.title("Number of matches per season")
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.countplot(x=matches['winner'], data=matches)
plt.title("Most win by a team")
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.countplot(x=matches['season'], data=matches, hue=matches['toss_decision'])
plt.title("Toss decision per season")
plt.show()
```

```python
pd.value_counts(matches['toss_decision'])
```

```python
parties = ['BJP','AAP','CONG','OTHERS']
votes = [300,100,50,82]
```

```python
plt.pie(votes,labels=parties,
        startangle=90,counterclock=False,
       autopct='%1.2f%%',colors=['green','cyan','orange','yellow'],
       explode=(0,0,0.4,0),shadow=True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,7))
axes = matches['season'].value_counts().plot.bar(width=0.9,color=sns.color_palette('inferno'))
```

```python
plt.figure(figsize=(10,7))
axes = matches['season'].value_counts().plot.bar(width=0.9,color=sns.color_palette('inferno'))
for patch in axes.patches:
    axes.annotate(format(patch.get_height()),(patch.get_x() + 0.3, patch.get_height() + 0.3))
```
